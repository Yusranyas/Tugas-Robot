```python
from google.colab import files #mengimpor modul dari google colab
import cv2 #mengimpor OpenCV
import numpy as np # mengimpor numpy
import matplotlib.pyplot as plt # mengimpor matplotlib
#Unggah gambar
uploaded = files.upload() #untuk membuka dialog unggah file
img_path = list(uploaded.keys())[0] #mengambil file yg diunggah
img = cv2.imread(img_path, 0) # Baca gambar menggunakan fungsi cv2.imread
#Buat kernel moving average 3x3
kernel = np.ones ((3, 3), np.float32) / 9 #membuat kernel 3x3 dengan semua elemen bernilai 1/9
#Terapkan filter
filtered_img = cv2.filter2D(img, -1, kernel) #menerapkan filter konvolusi pada gambar
#Tampilkan hasil
plt.figure(figsize=(10, 5)) #membuat figure baru dengan ukuran 10x5
plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original Image') #menampilkan gambar pada subplot pertama
plt.subplot(1, 2, 2), plt.imshow(filtered_img, cmap='gray'), plt.title('Filtered Image') #menampilkan gambar yang telah di filter pada subplot kedua
plt.show() #menampilkan plot
```

Setelah menerapkan kernel moving average 3x3, setiap piksel dalam gambar dihasilkan dari rata-rata nilai piksel di sekitarnya. Proses ini dikenal sebagai smoothing atau perataan.

```python
from google.colab import files #mengimpor modul
import cv2 # mengimpor modul
import matplotlib.pyplot as plt #mengimpor modul
#Unggah gambar
uploaded = files.upload() #untuk membuka dialog unggah file
img_path = list(uploaded.keys())[0] #mengambil file yg diunggah
img = cv2.imread(img_path, 0) # Baca gambar menggunakan fungsi cv2.imread
#Buat SIFT detector
sift = cv2.SIFT_create() #membuat objek SIFT
#Deteksi keypoints dan deskriptor
keypoints, descriptors = sift.detectAndCompute(img, None) #untuk mendeteksi keypoints pada gambar
#Gambar keypoints
img_with_keypoints = cv2.drawKeypoints(img, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS) #menggambar keypoints pada gambar input
#Tampilkan hasil
plt.figure(figsize=(10, 5)) #membuat figure baru dengan ukuran 10x5
plt.imshow(img_with_keypoints) #menampilkan gambar dengan keypoints
plt.title('SIFT Keypoints') #menambah judul pada plot
plt.axis('off') #menghilangkan sumbu gambar
plt.show() #menampilkan plot
```

Keypoints ditampilkan sebagai lingkaran di atas gambar asli. Setiap lingkaran menunjukkan area unik pada gambar yang terdeteksi oleh algoritma SIFT. Ukuran lingkaran merepresentasikan skala fitur, yaitu seberapa besar area lokal yang memengaruhi keypoint tersebut. Arah garis dalam lingkaran menunjukkan orientasi fitur lokal, yang digunakan untuk memastikan invariansi terhadap rotasi.

```python
from google.colab import files #mengimpor modul
import cv2 #mengimpor cv2
import matplotlib.pyplot as plt #mengimpor matplotlib
#Unggah gambar
uploaded = files.upload() #untuk membuka dialog unggah file
img_path = list(uploaded.keys())[0] #mengambil file yg diunggah
img = cv2.imread(img_path) # Baca gambar menggunakan fungsi cv2.imread
#Pisahkan channel warna dan buat histogram
colors = ('b', 'g', 'r') #mendefinisikan nama nama channel warna
plt.figure(figsize=(8, 5)) #membuat figure baru ukuran 8x5
for i, color in enumerate (colors): #melakukan pengulangan untuk setiap channel berwarna b,g,r
  histogram = cv2.calcHist([img], [i], None, [256], [0, 256]) #menghitung histogram untuk setiap warna
  plt.plot(histogram, color=color) #memplot histogram untuk channel warna saat ini dengan warna yg sesuai
  plt.xlim([0, 256]) #mengatur rentang sumbu x dari 0 sampai 256
plt.title('Color Histogram') #menmbah judul pada grafik
plt.xlabel('Pixel Value') #menambah label pada x
plt.ylabel('Frequency') #menambah label pada y
plt.show() #menampilkan grafik
```

Grafik menunjukkan distribusi nilai piksel untuk setiap channel warna (biru, hijau, dan merah) dalam gambar. Sumbu x menunjukkan nilai intensitas piksel (0 hingga 255). Sumbu y menunjukkan jumlah piksel untuk setiap nilai intensitas.

```python
from google.colab import files #mengimpor modul
import cv2 #mengimpor cv2
import matplotlib.pyplot as plt #mengimpor matplotlib
#Unggah gambar
uploaded = files.upload() #untuk membuka dialog unggah file
img_path = list(uploaded.keys())[0] #mengambil file yg diunggah
img = cv2.imread(img_path, 0) # Baca gambar dalam grayscale
#Terapkan Sobel filter
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3) # Tepi horizontal
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3) # Tepi vertikal
sobel_combined = cv2.magnitude (sobel_x, sobel_y) #menggabungkan horizontal dan vertikal

#Tampilkan hasil
plt.figure(figsize=(15, 5)) #membuat figure baru 15x5
plt.subplot(1, 3, 1), plt.imshow(sobel_x, cmap='gray'), plt.title('Sobel X') #menampilkan hasil sobel filter horizontal
plt.subplot(1, 3, 2), plt.imshow(sobel_y, cmap='gray'), plt.title('Sobel Y') #menampilkan hasil sobel filter vertikal
plt.subplot(1, 3, 3), plt.imshow(sobel_combined, cmap='gray'), plt.title('Combined Sobel') #menampilkan gabungan horizontal dan vertikal
plt.show() #menampilkan semua subplot
```

sobel x : Menonjolkan tepi yang tegak lurus terhadap sumbu x (horizontal). Tepi vertikal, seperti sisi objek atau garis-garis vertikal, terlihat lebih jelas. Area gambar dengan perubahan intensitas besar dalam arah horizontal memiliki nilai gradien tinggi.
sobel y : Menonjolkan tepi yang tegak lurus terhadap sumbu y (vertikal). Tepi horizontal, seperti garis-garis mendatar atau batas objek horizontal, lebih menonjol. Area dengan perubahan intensitas besar dalam arah vertikal memiliki nilai gradien tinggi.
combined sobel : Menggabungkan informasi dari gradien horizontal dan vertikal. Memberikan peta tepi lengkap dengan tepi yang lebih menonjol di seluruh arah. Menghasilkan gambaran yang lebih detail tentang struktur tepi dalam gambar.

```python
from google.colab import files #mengimpor modul
import cv2 #mengimpor cv2
import matplotlib.pyplot as plt #mengimpor matplotlib
from skimage.feature import hog #mengimpor hog dari skimage
import numpy as np #mengimpor numpy
#Unggah gambar
uploaded = files.upload() #untuk membuka dialog unggah file
img_path = list(uploaded.keys())[0] #mengambil file yg diunggah
img = cv2.imread(img_path, 0) # Baca gambar dalam grayscale
# Ekstraksi fitur HOG dengan parameter yang disesuaikan
features, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), #menghitung hog dan menghasilkan representasi
                          cells_per_block=(2, 2), visualize=True, channel_axis=None)
#Normalisasi HOG image dengan scaling manual jika normalisasi default gagal
hog_image_normalized = (hog_image - hog_image.min()) / (hog_image.max() - hog_image.min()) * 255 #normalisasi nilai intensitas pada hog_image ke skala 0-255
hog_image_normalized = np.uint8(hog_image_normalized) # Konversi ke uint8 untuk visualisasi
#Tampilkan hasil
plt.figure(figsize=(10, 5)) #membuat figure 10x5
plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original Image') #menampilkan gambar asli
plt.subplot(1, 2, 2), plt.imshow(hog_image_normalized, cmap='gray'), plt.title('HOG Representation') #menampilkan representasi hog
plt.show() #menampilkan kedua subplot
```

Menampilkan informasi arah gradien di seluruh gambar. Setiap sel dalam blok HOG berisi histogram yang menggambarkan distribusi arah gradien. Garis-garis kecil dalam representasi HOG menunjukkan arah dominan gradien pada area tertentu.
